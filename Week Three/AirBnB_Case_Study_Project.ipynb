```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv(r'C:\Users\adoro\OneDrive\Desktop\CT Python\Week Three\Assignments\AB_NYC_2019.csv')
df
```

```python
# Which hosts are the busiest and why?

df.groupby(['host_name']).count().sort_values(['availability_365'], ascending = False)[['availability_365', 'reviews_per_month','calculated_host_listings_count']]

```

To determine which hosts were the busiest, we wanted to focus on the availability of the listings, the number of listings, and the number of reviews the hosts receive. Based on the information above, we believe that John is the busiest as he has the highest review per availability while having a top 5 amount of listings.

```python
# How many neighborhood groups are available and which shows up the most?
df.groupby(['neighbourhood_group']).count().sort_values(['neighbourhood_group'], ascending = True)
```

Based on our review, there are 5 neighborhood groups within New York: Bronx, Brooklyn, Manhattan, Queens, and Staten Island. Manhattan was the one that showed up the most with 21,661 listings. The 21,661 listings were found under the id number as all listings will have an unique number.

```python
# Are private rooms the most popular in manhattan?
df.groupby(['neighbourhood_group', 'room_type']).count().sort_values(['neighbourhood_group'])[['calculated_host_listings_count']]
```

Based on our review, entire home/apt are the most popular in Manhattan with 13,199 listings.
Privates rooms were second in popularity in Manhattan with 7982 listings.

```python
# Which hosts are the busiest and based on their reviews?

df.groupby(['host_name']).count().sort_values(['reviews_per_month'], ascending = False)[['calculated_host_listings_count', 'reviews_per_month', 'availability_365']]
```

Similar to our review above, we believe John was the busiest per capita of reviews per month to calculated host listings. The second busiest was Alex as he was second in reviews per listings.

```python
#Which neighorhood group has the highest average price?
df.groupby(['neighbourhood_group']).mean().sort_values(['price'], ascending = False)[['price']].round(2)
```

Based on our findings by filtering the data to show neighbourhood groups and average price, Manhattan had the highest average price of $196.88.

```python
# Which neighbor hood group has the highest total price?
df.groupby(['neighbourhood_group']).max().sort_values(['price'], ascending = False)[['price']]
```

In preparation for this question, we filtered the data by neighbourhood group and price and then selected the top 5 maximum amounts. Our analysis shows that Brooklyn, Manhattan, and Queens each charge a high price of $10,000.

```python
#Which top 5 hosts have the highest total price?
df.groupby(['host_name']).max().sort_values(['price'], ascending = False)[['price']].head(5)
```

After filtering the data by host name and price, we have found that Kathrine, Erin and Jelena charge a price of $10,000 and Amy/Olson charge a price of $9999.

```python
# Who currently has no (zero) availability with a review count of 100 or more?


conditional = (df['number_of_reviews'] >= 100) & (df['availability_365'] == 0)
df[conditional].groupby(['host_name']).count().sort_values(['number_of_reviews'], ascending = False)[['number_of_reviews', 'availability_365']]


```

By filtering the data, we have found that 141 hosts have met the requirements of having zero availability and at least 100 reviews.

```python
# What host has the highest total of prices and where are they located?
df.groupby(['host_name']).max().sort_values(['price'], ascending = False)[['price', 'neighbourhood_group', 'neighbourhood']].head(5)
```

The highest total of prices through our analysis are currently located in Kips Bay, Queens; Williamsburg, Queens; and Williamsburg, Manhattan.

```python
# When did Danielle from Queens last receive a review?
last_review = (df['host_name'] == 'Danielle') & (df['neighbourhood_group'] == 'Queens')

df[last_review].groupby(['host_name']).max().sort_values(['neighbourhood_group'], ascending = False)['last_review']
```

Through all of Danielle (from Queen's) listings, her last review came on 7/8/2019.
